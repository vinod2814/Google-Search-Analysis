```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from pytrends.request import TrendReq
trend = TrendReq()
```

```python
trend
```

create a DataFrame of the top 20 countries which search for “Machine Learning” on Google:



```python
trend.build_payload(kw_list=["Machine Learning"])
data = trend.interest_by_region()
data = data.sort_values(by="Machine Learning" ,ascending=False)
data = data.head(20)
print(data)
```

```python
# Python
trend.build_payload(kw_list=["Python"])
data = trend.interest_by_region()
data = data.sort_values(by="Python" ,ascending=False)
data = data.head(20)
print(data)
```

So, according to the above results, the search queries based on “Machine learning” and "Python" are mostly done in China.

```python
# bar plot for Machine Learning
data.reset_index().plot(x="geoName",y="Machine Learning",figsize=(15,8),kind="bar")
```

```python
# bar plot for Python
data.reset_index().plot(x="geoName",y="Python",figsize=(15,8),kind="bar")
```

let’s have a look at the trend of searches to see how the total search queries based on “Machine Learning” and "Python" increased or decreased on Google:

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Machine Learning'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(20, 15))
data['Machine Learning'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Machine Learning', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()

```

```python
# Python Trends
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Python'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(20, 15))
data['Python'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Python', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()

```
